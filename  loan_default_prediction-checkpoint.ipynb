```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(
x, y, 
test_size=0.2,
random_state=42,
stratify=y)
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=42)
x_train_smote, y_train_smote=smote.fit_resample(x_train,y_train)
```

```python
print(y_train_smote.value_counts())
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(random_state=42,
                         max_iter=1000)
model.fit(x_train_smote, y_train_smote)
model_pred=model.predict(x_test)
model_prob=model.predict_proba(x_test)[:,1]
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score,f1_score,roc_auc_score
def evaluate_model(model,y_true,y_pred,y_prob):
    print(f"\n{model}")
    print("-"*30)
    print("Accuracy :", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred))
    print("Recall   :", recall_score(y_true, y_pred))
    print("F1 Score :", f1_score(y_true, y_pred))
    print("ROC AUC  :", roc_auc_score(y_true, y_prob))
    
```

```python
evaluate_model("LogisticRegression",y_test, model_pred, model_prob)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(max_depth=8,
                         min_samples_split=10,
                         min_samples_leaf=5,
                         random_state=42)
dt.fit(x_train_smote, y_train_smote)
dt_pred=dt.predict(x_test)
dt_prob=dt.predict_proba(x_test)[:,1]                           
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score,f1_score,roc_auc_score
def evaluate_model(dt,y_true,y_pred,y_prob):
    print(f"\n{dt}")
    print("-"*30)
    print("Accuracy :", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred))
    print("Recall   :", recall_score(y_true, y_pred))
    print("F1 Score :", f1_score(y_true, y_pred))
    print("ROC AUC  :", roc_auc_score(y_true, y_prob))
    
```

```python
evaluate_model("DecisionTreeClassifier",y_test, dt_pred, dt_prob)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=300,
                          max_depth=10,
                          min_samples_split=10,
                          min_samples_leaf=5,
                          n_jobs=-1,
                          random_state=42
                         )
rf.fit(x_train_smote, y_train_smote)
rf_pred=rf.predict(x_test)
rf_prob=rf.predict_proba(x_test)[:,1]                           
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score,f1_score,roc_auc_score
def evaluate_model(rf,y_true,y_pred,y_prob):
    print(f"\n{rf}")
    print("-"*30)
    print("Accuracy :", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred))
    print("Recall   :", recall_score(y_true, y_pred))
    print("F1 Score :", f1_score(y_true, y_pred))
    print("ROC AUC  :", roc_auc_score(y_true, y_prob))
    
```

```python
evaluate_model("RandomForestClassifier",y_test, rf_pred, rf_prob)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Logistic Regression
fpr_lr, tpr_lr, _ = roc_curve(y_test, model_prob)
auc_lr = roc_auc_score(y_test, model_prob)

# Decision Tree
fpr_dt, tpr_dt, _ = roc_curve(y_test, dt_prob)
auc_dt = roc_auc_score(y_test, dt_prob)

# Random Forest
fpr_rf, tpr_rf, _ = roc_curve(y_test, rf_prob)
auc_rf = roc_auc_score(y_test, rf_prob)

# XGBoost
fpr_xgb, tpr_xgb, _ = roc_curve(y_test, xgb_prob)
auc_xgb = roc_auc_score(y_test, xgb_prob)

# Plot ROC Curves
plt.figure(figsize=(8,6))

plt.plot(fpr_lr, tpr_lr, label=f'Logistic Regression (AUC = {auc_lr:.3f})')
plt.plot(fpr_dt, tpr_dt, label=f'Decision Tree (AUC = {auc_dt:.3f})')
plt.plot(fpr_rf, tpr_rf, label=f'Random Forest (AUC = {auc_rf:.3f})')
plt.plot(fpr_xgb, tpr_xgb, label=f'XGBoost (AUC = {auc_xgb:.3f})')

# Random classifier
plt.plot([0,1], [0,1], 'k--', label='Random Guess')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Comparison')
plt.legend(loc='lower right')
plt.grid(True)

plt.show()
```

The ROC curves demonstrates that Logistic regression had the strongest discriminative ability followed closely by XGBoost 
although XGBoost produced the highest overall accuracy, Logistic regression provided better separation between default and non 
default borrowers.
